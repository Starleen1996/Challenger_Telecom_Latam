```python
# Vamos a limpiar algunos datos en la columna "account_charges_total" y convertir la columna "account_charges_total" a float64

# Reemplaza espacios vac铆os o cadenas vac铆as con NaN
df_clientes_LATAM['account_charges_total'] = df_clientes_LATAM['account_charges_total'].astype(str).str.strip()
df_clientes_LATAM['account_charges_total'] = df_clientes_LATAM['account_charges_total'].replace('', np.nan)

# Ahora convierte a float
df_clientes_LATAM['account_charges_total'] = df_clientes_LATAM['account_charges_total'].astype(np.float64)
```

## Distribuci贸n de evasi贸n
En este paso, el objetivo es comprender c贸mo est谩 distribuida la variable "churn" (evasi贸n) entre los clientes. Utiliza gr谩ficos para visualizar la proporci贸n de clientes que permanecieron y los que se dieron de baja.

```python
import matplotlib.pyplot as plt

labels = ['No Vigente', 'Vigente']
sizes = df_clientes_LATAM['cliente_vigente'].value_counts().sort_index() #Cuenta cu谩ntos clientes hay en cada categor铆a (0 = No vigente, 1 = Vigente).
#sort_index() asegura que el orden sea [0, 1], lo cual coincide con el orden de labels.

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['red', 'mediumseagreen']) #plt.pie(...) : Crea un gr谩fico de torta
# sizes: los valores a graficar (cantidades de cada grupo). labels: nombres que acompa帽an cada secci贸n. autopct='%1.1f%%': muestra el porcentaje con 1 decimal.
#startangle=90: empieza el gr谩fico desde la parte superior (mejora est茅tica). colors=['red', 'mediumseagreen']: color rojo para "no vigente" y verde para "vigente".

plt.title('Porcentaje de Clientes Vigentes vs No Vigentes')
plt.axis('equal')  # Para que sea un c铆rculo, Asegura que el gr谩fico sea un c铆rculo perfecto (no una elipse).
plt.show()

# Evidenciamos que el 73.5% de los clientes se han retidado de la compa帽ia , es decir no estan vigentes actualmente.
```
