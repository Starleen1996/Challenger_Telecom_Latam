#### Manejo de inconsistencias
Ahora que has identificado las inconsistencias, es momento de aplicar las correcciones necesarias. Ajusta los datos para asegurarte de que est茅n completos y coherentes, prepar谩ndolos para las siguientes etapas del an谩lisis.

```python
df_clientes_LATAM.columns
```

```python
# Vamos a estandarizar los nombres de las columnas y las vamos a colocar todas en minusculas

df_clientes_LATAM = df_clientes_LATAM.rename(columns=lambda x: x.lower())
df_clientes_LATAM.columns
```

```python
# Tambien remplazamos la separaci贸n por la columnas con "_"
df_clientes_LATAM.columns = df_clientes_LATAM.columns.str.lower().str.replace('.', '_')
df_clientes_LATAM.columns
```

```python
df_clientes_LATAM['customerid'].unique()
```

```python
import numpy as np
```

```python
# Vamos a limpiar algunos datos en la columna "account_charges_total" y convertir la columna "account_charges_total" a float64

# Reemplaza espacios vac铆os o cadenas vac铆as con NaN
df_clientes_LATAM['account_charges_total'] = df_clientes_LATAM['account_charges_total'].astype(str).str.strip()
df_clientes_LATAM['account_charges_total'] = df_clientes_LATAM['account_charges_total'].replace('', np.nan)

# Ahora convierte a float
df_clientes_LATAM['account_charges_total'] = df_clientes_LATAM['account_charges_total'].astype(np.float64)
```

```python
df_clientes_LATAM['account_charges_total'].dtype
```

```python
# Vamos a eliminar las filas vacias de la columna churn


# Contar las filas antes de eliminar
filas_antes = len(df_clientes_LATAM)

# Limpiar espacios y convertir vac铆os a NaN en la columna 'churn'
df_clientes_LATAM['churn'] = df_clientes_LATAM['churn'].astype(str).str.strip()
df_clientes_LATAM['churn'].replace('', np.nan, inplace=True)

# Eliminar filas con NaN en 'churn'
df_clientes_LATAM = df_clientes_LATAM.dropna(subset=['churn'])

# Contar las filas despu茅s de eliminar
filas_despues = len(df_clientes_LATAM)

# Imprimir cu谩ntas filas fueron eliminadas
print(f"Se eliminaron {filas_antes - filas_despues} filas vac铆as o nulas en la columna 'churn'.")
```

```python
# Verificar espacios vacios o en blanco despues de limpiar los datos
#sirve para contar cu谩ntos valores vac铆os (cadenas vac铆as o espacios en blanco) hay en cada columna de un DataFrame llamado df_clientes_LATAM.
df_clientes_LATAM.apply(lambda x: x.astype(str).str.strip() == '').sum()
```
